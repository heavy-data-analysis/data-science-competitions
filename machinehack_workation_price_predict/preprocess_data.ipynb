### Travel Date

### Itinerary

```python
df['Itinerary']
```

```python
nights = []
places = []
for i in range(0,len(df['Itinerary'])):
    arr = df['Itinerary'][i].split('.')
    p_cnt = 0
    n_cnt = 0
    for j in range(0, len(arr)):
        p_cnt = p_cnt + 1
        arr[j] = arr[j].strip()
        n_cnt = n_cnt + int(arr[j][0])
    #print(arr)
    nights.append(n_cnt)
    places.append(p_cnt)
```

```python
df['total_N'] = nights
df['total_P'] = places
```

```python
nights = []
places = []
for i in range(0,len(df_test['Itinerary'])):
    arr = df_test['Itinerary'][i].split('.')
    p_cnt = 0
    n_cnt = 0
    for j in range(0, len(arr)):
        p_cnt = p_cnt + 1
        arr[j] = arr[j].strip()
        n_cnt = n_cnt + int(arr[j][0])
    #print(arr)
    nights.append(n_cnt)
    places.append(p_cnt)
```

```python
df_test['total_N'] = nights
df_test['total_P'] = places
```

### Airline

```python
df['Airline'].unique()
```

```python
airlines = []
for i in range(0,len(df['Airline'])):
    arr = df['Airline'][i].split('|')
    air_set = {""}
    for j in range(0, len(arr)):
        arr[j] = arr[j].strip()
        air_set.add(arr[j])
    air_set.remove('')
    arr = []
    for x in air_set:
        arr.append(x)
    arr.sort()
    ans = ""
    for i in range(len(arr)):
        if i != 0:
            ans = ans + '|' + arr[i]
        else:
            ans = arr[i]
    airlines.append(ans)
df['Airline'] = airlines
```

```python
airlines = []
for i in range(0,len(df_test['Airline'])):
    arr = df_test['Airline'][i].split('|')
    air_set = {""}
    for j in range(0, len(arr)):
        arr[j] = arr[j].strip()
        air_set.add(arr[j])
    air_set.remove('')
    arr = []
    for x in air_set:
        arr.append(x)
    arr.sort()
    ans = ""
    for i in range(len(arr)):
        if i != 0:
            ans = ans + '|' + arr[i]
        else:
            ans = arr[i]
    airlines.append(ans)
df_test['Airline'] = airlines
```
