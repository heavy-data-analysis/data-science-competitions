```python
import pandas as pd

from pycaret.regression import *

import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('../data/feat_eng_train1.csv')
test = pd.read_csv('../data/feat_eng_test1.csv')
```

```python
data_setup = setup(
    data=train,
    target='tumor_size',
    preprocess=False,
    fold_shuffle=True,
)
```

```python
best_model = compare_models(sort='RMSE')
```

```python
evaluate_model(best_model)
```

```python
final_model = finalize_model(best_model)
```

```python
preds = predict_model(final_model, data=test)
```

```python
print(preds.shape)
preds.head()
```

```python
sub = pd.DataFrame({ 'tumor_size': preds['Label'] })
print(sub.shape)
sub.head()
```

```python
sub.to_csv('../submissions/pycaret1.csv', index=False)
```
